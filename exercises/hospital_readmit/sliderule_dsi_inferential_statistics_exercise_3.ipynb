# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

### Part B of the Exercise:
We will do a hypothesis test, eleminating the common samples from both the tests.

Here are our hypotheses:

H0 : The null hypothesis is that the excess readmission rate is uncorrelated with number of discharges

HA : The alternate hypothesis is that excess readmission rate is negatively correlated with number of discharges.

Let us have alpha = 0.01

Stated in plain English, we need to check if excess readmission rate for small hospitals and large hospitals come from same population (H0) or not (HA).

We will conduct a two sample hypothesis testing

```python
aub = set(list(df1000['Hospital Name'].unique()) + list(df100['Hospital Name'].unique()))
amb  = set(list(df1000['Hospital Name'].unique())) - set(list(df100['Hospital Name'].unique()))
bma  = set(list(df100['Hospital Name'].unique())) - set(list(df1000['Hospital Name'].unique()))
acb = aub - amb - bma

#Cross Verify
print('Cardinality of acb',len(acb))
acb
```

Note that amb and bma are the samples of our interest to conduct two sample hypothesis tests. amb has hospitals with > 1000 discharges and not any < 100 discharges, bma has hospitals with < 100 discharges and not any > 1000 discharges. One needs to pull out the records of only these hospitals.

```python
df1000['Hospital Name'] = df1000['Hospital Name'].apply(lambda x: x if x in amb else '') 
```

```python
df100['Hospital Name'] = df100['Hospital Name'].apply(lambda x: x if x in bma else '') 
```

```python
df100 = df100[df100['Hospital Name'] != '']
df1000 = df1000[df1000['Hospital Name'] != '']
```

The above also makes no sense. Hence we can not conduct a two sample test. But let's say we go ahead blindly, as required by the exercise. Let's take < 350 and > 850 samples

Assumption verification:
a) Records are independent
b) Sample size are much greater than 30. The distributions are nearly normal
Less than 10% of population
 Central limit theorem can be applied. 

```python
df100[['Excess Readmission Ratio','Number of Discharges']].describe()
```

```python
df1000[['Excess Readmission Ratio','Number of Discharges']].describe()
```

```python
mean1 = df100['Excess Readmission Ratio'].mean()
mean2 = df1000['Excess Readmission Ratio'].mean()
var1 = df100['Excess Readmission Ratio'].var()
var2 = df1000['Excess Readmission Ratio'].var()
n1 = df100['Excess Readmission Ratio'].count()
n2 = df1000['Excess Readmission Ratio'].count()
arr = np.round(np.array([mean1,mean2,var1,var2]),4)
print(arr)
```

```python
zscore = (mean1 - mean2)/np.sqrt((var1/n1) + (var2/n2))
print("Z-score:",round(zscore,2))
```

**Since Z-score is so high at 7.47, the p-value would be nearly 0, we reject the null hypothesis in favor of alternate hypothesis**
